### Basic Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('data/bankloan.csv')

# Basic Info
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Unique values in categorical columns
categorical_cols = ['ZIP.Code', 'Family', 'Education', 'Personal.Loan',
                    'Securities.Account', 'CD.Account', 'Online', 'CreditCard']
print("\nUnique values in categorical columns:")
for col in categorical_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage', 'Personal.Loan']], hue='Personal.Loan')
plt.show()
```

```python
# Count plots for categorical columns
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Distribution plots for continuous variables
continuous_cols = ['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage']
for col in continuous_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```
