# Cleaning the Data

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

```

```python
# Drop ID and ZipCode (not useful for prediction)
df.drop(['ID', 'ZIP.Code'], axis=1, inplace=True)
```

```python
# Features and Target
X = df.drop('Personal.Loan', axis=1)
y = df['Personal.Loan']
```

```python
# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Save the fitted scaler - to use for streamlit UI
import joblib
joblib.dump(scaler, "scaler.joblib")
```

```python
# Split into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=10, stratify=y)
```

```python
print("Data is ready.")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train distribution:\n", y_train.value_counts(normalize=True))
```

```python
import keras_tuner as kt
from tensorflow.keras import layers, models
```

```python
def build_model(hp):
    model = models.Sequential()
    model.add(layers.Input(shape=(X_train.shape[1],)))

    # Tune number of layers
    for i in range(hp.Int("num_layers", 1, 3)):
        model.add(layers.Dense(
            units=hp.Int(f"units_{i}", min_value=16, max_value=128, step=16),
            activation=hp.Choice("activation", ["relu", "tanh"])
        ))
        if hp.Boolean(f"dropout_{i}"):
            model.add(layers.Dropout(rate=hp.Float(f"dropout_rate_{i}", 0.1, 0.5, step=0.1)))

    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=hp.Choice("optimizer", ["adam", "rmsprop", "sgd"]),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC()]
    )
    return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective="val_accuracy",
    max_trials=10,
    executions_per_trial=1,
    directory="loan_tuning",
    project_name="bankloan_model"
)

tuner.search(X_train, y_train, epochs=20, validation_split=0.2)
```

```python
tuner.search(X_train, y_train, epochs=20, validation_split=0.2)
```

```python
# View best model
best_hps = tuner.get_best_hyperparameters(1)[0]
print("Best hyperparameters:")
print(best_hps.values)
```

# Feed Forward Neural Network


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping  # <--- Add this line
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix


# Define EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Build model with best hyperparameters from tuner
best_model = tuner.hypermodel.build(best_hps)


# Define EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Train model on training data with validation split and early stopping
history = best_model.fit(
    X_train, y_train,
    epochs=20,
    validation_split=0.2,
    callbacks=[early_stopping]
)
```

```python
# Save the trained model
best_model.save("bankloan_model_tuned.keras")

# Evaluate model performance on test data
loss, accuracy, auc = best_model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
print(f"Test AUC: {auc:.4f}")

# Predict probabilities and convert to class labels (threshold 0.5)
y_pred_prob = best_model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Print classification report and confusion matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

```

```python
# Plot training & validation accuracy
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.legend()
plt.show()

# Plot training & validation loss
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Model Loss')
plt.legend()
plt.show()
```

# RNN Model

```python
from tensorflow.keras.layers import SimpleRNN, Reshape
```

```python

X_train_rnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_rnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
# Build RNN model
rnn_model = Sequential([
    SimpleRNN(64, activation='relu', input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.1),
    Dense(1, activation='sigmoid')
])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
# Compile the model
rnn_model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])


# Define EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Train the RNN model
history_RNN =rnn_model.fit(X_train_rnn, y_train, epochs=20, batch_size=32, validation_split=0.2, callbacks=[early_stopping], verbose=1)

# Evaluate RNN model
rnn_loss, rnn_acc = rnn_model.evaluate(X_test_rnn, y_test.to_numpy(), verbose=0)
print(f"RNN Test Accuracy: {rnn_acc:.4f}")
```

```python
# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history_RNN.history['accuracy'], label='Train Accuracy')
plt.plot(history_RNN.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history_RNN.history['loss'], label='Train Loss')
plt.plot(history_RNN.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

# LSTM Model

```python
from tensorflow.keras.layers import LSTM
```

```python
# Build LSTM model
lstm_model = Sequential([
    LSTM(64, activation='tanh', input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.1),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
lstm_model.compile(optimizer='adam',
                   loss='binary_crossentropy',
                   metrics=['accuracy'])
```

```python
from sklearn.utils.class_weight import compute_class_weight


# Compute class weights
classes = np.unique(y_train)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weight_dict = dict(zip(classes, class_weights))
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

lstm_history = lstm_model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    class_weight=class_weight_dict,
    callbacks=[early_stopping],
    verbose=1
)
```

### This will stop training if validation loss doesn't improve for 5 epochs and restore the best weights.

```python
# Evaluate LSTM model
lstm_loss, lstm_accuracy = lstm_model.evaluate(X_test, y_test.to_numpy(), verbose=0)
print(f"LSTM Test Accuracy: {lstm_accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Plot loss
plt.subplot(1, 2, 1)
plt.plot(lstm_history.history['loss'], label='Train Loss')
plt.plot(lstm_history.history['val_loss'], label='Validation Loss')
plt.title('LSTM Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

# Plot accuracy
plt.subplot(1, 2, 2)
plt.plot(lstm_history.history['accuracy'], label='Train Accuracy')
plt.plot(lstm_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('LSTM Training Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

## Background and Problem Statement  
Loan approval is a critical decision-making process for financial institutions. Manual evaluation of applicants based on various factors such as credit history, income, loan amount, and demographic data can be inefficient and inconsistent. The goal of this project was to develop an AI-based system that can learn patterns from historical loan application data and accurately predict whether a new application should be approved.

Using deep learning techniques, we sought to build a model that can support automated, reliable loan approval predictions for structured (tabular) data, thereby assisting financial institutions in decision-making.

---

## Dataset  
The dataset was sourced from [Kaggle – Bank Loan Approval Dataset](https://www.kaggle.com/datasets/vikramamin/bank-loan-approval-lr-dt-rf-and-auc). It included structured data on previous loan applications, covering:

- Applicant’s income, education, and employment status  
- Loan amount and loan term  
- Credit history  
- Marital status, gender, and number of dependents  
- Approval status (target variable)

### Data Preprocessing Included:
- Handling missing values using imputation  
- One-hot encoding for categorical features  
- Normalization of numerical variables  
- Removal of outliers (where applicable)

---

## Model Architecture  
We implemented a fully connected deep neural network using the **TensorFlow/Keras** framework. The architecture included:

- Input layer for tabular features  
- Multiple hidden layers with ReLU activation  
- Dropout layers for regularization  
- Final output layer with sigmoid activation for binary classification (approved vs not approved)

### Techniques Explored:
- Dropout Regularization  
- Batch Normalization (tested but found to be ineffective for this dataset)  
- Model compiled with binary cross-entropy loss and Adam optimizer  

---

## Training & Evaluation  
We trained our model with an 80/20 training-validation split. Evaluation metrics used:

- Accuracy  
- Precision  
- Recall  
- F1-score  
- Confusion Matrix  

We tracked model performance using learning curves (loss vs epochs, accuracy vs epochs).

---

## Hyperparameter Tuning  
We experimented with various combinations of:

- Learning rate  
- Batch size  
- Number of hidden layers and units  
- Dropout rates  

The best-performing configuration was selected based on validation accuracy and generalization performance.

---

## Results and Observations  
The final model achieved satisfactory performance, accurately classifying most applications in the test set.

### Key Observations:
- Preprocessing played a crucial role in improving performance  
- Batch normalization did not provide a significant benefit for this shallow architecture and dataset size  
- Dropout improved generalization and helped reduce overfitting  

---
